## 2.2 Check for null values and modify the dataset

From the description of the columns above, the following heatmap gives us important information on how to deal with the null values:

* The purpose of this project is to use the urinary levels of certain biomarkers to detect presence of PDAC. This means we are not concerned (for now) with the stage of cancer or the type of benign disease.
    
    * *Create a new output variable called 'has_cancer'* with values 0 (no) and 1 (yes) using information from the columns 'diagnosis', 'stage', and 'benign_sample_diagnosis'


* The column 'plasma_CA19_9' indicates  blood plasma levels of a protein, and was only collected for some of the patients.
    
    * *Drop 'plasma_CA19_9'* since we are only interested in urine  biomarkers
    
    
* The column 'REG1A' was only collected in 306 of the 590 patients
    
    * *Drop 'REG1A'* for  now. This may be  revisited later for more domain specific analysis
    
    
* The  columns 'sample_id', 'patient_cohort', and 'sample_origin' do not give enough information without more context
    
    * *Drop these columns*
    
    
* The column 'sample_origin' is also not meaningful under the assumption that the same sample collection procedure was used in all centers
    
    * *Drop sample_origin*
    
    
* The column 'sex' is categorical, with two values M and F. 
    
    * *Convert to dummy variable for one of the genders*

```python
# Define function to clean data

def clean_data(data, cohort = 1):
    '''
    Takes the full dataset as an argument and returns the cleaned dataset
    Also takes an optional argument for cohort to decide how to create the target variables.
    
    Cohort values are as follows (Target variables 0 and 1 in order):
    1: Control+Benign (Target variable 0) vs. Cancer (Target variable 1)
    2: Control+Benign vs. Stage I/II cancer
    3: Control vs. Benign
    4: Benign vs. Stage I/II cancer
    
    Note: These cohorts have nothing to do with the 'cohort' column in the dataset
    '''
    
    # Deal with categorical variable 'sex'
    # Create dummy  variable and concatenate to dataframe
    
    is_male = pd.get_dummies(data['sex'], drop_first=True)
    data = pd.concat([data, is_male], axis=1)
    
    data_cleaned = data
    
    if cohort == 1:
        # Create output column 'has_cancer'

        data_cleaned['has_cancer'] = [1 if x == 3 else 0 for x in data['diagnosis']]

        
    elif cohort == 2:
        # Select required rows
        
        data_cleaned.drop(data_cleaned[(data_cleaned.stage == 'III') | (data_cleaned.stage == 'IV')].index, inplace=True)
        
        # Create output column 'has_cancer'

        data_cleaned['has_cancer'] = [1 if x == 3 else 0 for x in data['diagnosis']]
    
    
    elif  cohort == 3:
        # select required rows
        
        data_cleaned.drop(data_cleaned[(data_cleaned.stage.isnull() == False)].index, inplace=True)
        
        # Create output column 'is_benign'
        
        data_cleaned['is_benign'] = [1 if x == 2 else 0 for x in data['diagnosis']]
    
    
    elif cohort == 4:
        # Select required rows
        
        data_cleaned.drop(data_cleaned[(data_cleaned.stage == 'III') | (data_cleaned.stage == 'IV')].index, inplace=True)
        data_cleaned.drop(data_cleaned[data_cleaned.stage.isnull() & data_cleaned.benign_sample_diagnosis.isnull()].index, inplace=True)
        
        # Create output column 'has_cancer'
        
        data_cleaned['has_cancer'] = [1 if x == 3 else 0 for x in data['diagnosis']]
 
    
    # Drop columns that aren't useful to the analysis

    columns_to_drop = [
        'sample_id',
        'patient_cohort',
        'sex',
        'diagnosis',
        'stage',
        'benign_sample_diagnosis',
        'sample_origin',
        'plasma_CA19_9',
        'REG1A']

    data_cleaned = data.drop(columns_to_drop, axis=1)


        
    return data_cleaned
```

```python
df_clean = clean_data(df, 1)

x = df_clean.drop('has_cancer',  axis=1)
y = df_clean['has_cancer']
```

# 4. Machine Learning

## 4.1 Test Different Algorithms with Default Hyperparameters

Before we begin, the first thing we need to do is make train test split. This initial split will give us a training set and an 'unseen' test set for the final validationtion. This  train-test split will be stratified to ensure a similar distribution of cancer and control samples  in  the split. 

We will then use k fold cross validation on the training set to compare different algorithms, tune the hyperparameters of the top choices, and finally evaluate our results on the test set before deployment.

```python
from sklearn.model_selection import train_test_split

# Cross validation for hyperparameter tuning and algorithm selection
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RepeatedStratifiedKFold
from sklearn.preprocessing import MinMaxScaler, RobustScaler, StandardScaler

# For model evaluation
from sklearn.model_selection import cross_val_score
from sklearn.metrics import roc_auc_score, recall_score, f1_score, accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import plot_confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.33,  
                                                    stratify=y)
```

Now let's define a function that will evaluate a dictionary of models and return the average performance scores of choice after cross validation

```python
# Define a function to run cross validation on a given set of models

def get_results(training_data, training_labels, models_dict, metrics_of_interest):
    '''
    Takes the data, labels, a dictionary of classification models, and a list of metrics as the inputs 
    and prints the average accuracy with cross validation.
    '''
  
    # Convert the dictionary of algorithms to evaluate into a list
    
    models_list = [(name, model_object) for name, model_object in models_dict.items()]
        
    for model in models_list:
        
        # Define a pipeline object takes in a  list of tuples  as steps.
        # Each step contains a string name and an object of the step, like a transform or a model

        steps = [
            ('scaler', RobustScaler()),
            (model)
        ]

        # Initialize pipeline
        pipeline = Pipeline(steps=steps)
        
        # Define evaluation precedures using RepeatedStratifiedKFold
        # Repeated means the cross validation is repeated
        # Stratified means each fold in t he k fold cross validation has a similar distribution of samples as the main dataset
        # We will repeat the process 5 times and use 10 folds
        
        cross_val = RepeatedStratifiedKFold(n_splits=10,  n_repeats=5)
        
        # Evaluate the model
        # Inefficient way to evaluate multiple metrics:
        
        print(model[0],':')
        
        for metric in metrics_of_interest:
            
            scores = cross_val_score(pipeline, 
                                     training_data, 
                                     training_labels, 
                                     scoring=metric, 
                                     cv=cross_val, 
                                     n_jobs=-1)
            
            print('Mean',metric, ':  %.3f (+/-%.3f)' % (np.mean(scores), np.std(scores)))
        
        print('\n')
```

Now try out some algorithms with default hyperparameters:

```python
# All algorithms we will try

from sklearn.linear_model  import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
```

```python
models_used = {
    'Logistic Regression': LogisticRegression(),
    'Random Forest': RandomForestClassifier(),
    'Support Vector Machines': SVC(),              
    'k-Nearest Neighbors': KNeighborsClassifier(n_neighbors=1)
}

metrics_used = [
    'accuracy',
    'recall',
    'f1'
]

get_results(x_train, y_train, models_used, metrics_used)
```

**Random forest** performs best _most  often_ in all metrics tested. 
Depending on the train test split, **SVM** _sometimes_ outperforms random forest in overall accuracy but not recall.

Note that accuracy isn't the most useful metric here. Since the application of this project is for medical diagnosis, we are interested in maximizing the true positives and minimizing the false negatives. Thus, **recall** is a good metric to use.

Precision gives a measure of the rate of false positives. Although this metric is not as important, we can still use an **F1 score** to get evaluate performance with respect to both recall and precision. Random forest consistently provides the highest average recall score which is out most important metric.

## 4.3 Hyperparameter Tuning

We will tune the random forest model to ensure optimized performance

```python
# Define a function to tune the hyperparameters

def tune_hyperparameters(training_data, training_labels, model, param_space):
    '''
    Function takes the data, labels, an algorithm, and the hyperparameter space 
    and returns the tuned model fit to the training data
    '''
    
    # Steps in pipeline
    steps = [
        ('scaler', RobustScaler()),
        ('model', model)
    ]
    
    
    # Initialize pipeline
    pipeline = Pipeline(steps=steps)
    
    random_search = RandomizedSearchCV(pipeline,
                            param_distributions=param_space,
                            scoring='recall',
                            n_jobs=-1,
                            n_iter=20,
                            cv=5)

    random_search.fit(training_data, training_labels)
    
    return random_search.best_estimator_
```

```python
# Define the hyperparameter spaces
# Note the formatting of the dictionary to account for the structure of the pipeline used in the function above

param_space_rf = {
    'model__max_depth': np.arange(1,20),
    'model__n_estimators': np.arange(0,1500,100),
    'model__min_samples_leaf': np.arange(1,20),
    'model__criterion': ['gini','entropy']
}


# Hyperparameter spaces for additional algorithms if we want to use them in the future 

param_space_svm = {
    'model__C': np.power(10., np.arange(-4,4)),
    'model__gamma': np.power(10., np.arange(-5,0))
}

param_space_knn = {
    'model__n_neighbors': np.arange(1,25)
}
```

```python
rf_tuned = tune_hyperparameters(x_train, y_train, RandomForestClassifier(), param_space_rf)
```

```python
print(rf_tuned)
```

## 4.4 Final Evaluation

Finally take the tuned model and predict on the proxy for the 'unseen' data to confirm satisfactory performance of model

```python
# Define a function to display the confusion matrix

def show_conf_mat(trained_model, label_names, title, fig_no, test_data = x_test, test_labels = y_test, labels=[1,0]):
    '''
    Takes in order, a trained model, a list of target class variables (labels), a list of label names, 
    and a string for the title of the confusion matrix. 
    Also takes in an integer value fig_no for the name of the saved image
    Test sets are x_test and y_test by default.
    Labels are [1,0] by default, 1 referring to the more serious condition
    '''
    y_pred = trained_model.predict(x_test)
    plot_confusion_matrix(trained_model, x_test, y_test, labels=labels, display_labels=label_names, colorbar=False, cmap='Greys')
    print("\nFinal Accuracy Score: %.3f" % accuracy_score(y_test, y_pred))
    print("Final Recall Score: %.3f" % recall_score(y_test, y_pred))
    print("Final F1 Score: %.3f" % f1_score(y_test, y_pred))
    print("\n", title)
    plt.savefig(r'C:\Users\sriha\OneDrive\Desktop\DS\My work\pancreatic-cancer-2\Figures\Fig %s.png' % fig_no)
```

# 5. Other Considerations

The goal here is to enable early detection of pancreatic cancer, so we should evaluate whether an algorithm can correctly identify cases of early stage PDAC from control or benign samples.

## 5.1 Compare 'No Cancer' vs. 'Stage I and II Cancer' Only

Note that 'No cancer' includes both control samples as well as benign pancreatic disease.

```python
df_no_vs_early = clean_data(df, 2)

x = df_no_vs_early.drop('has_cancer',  axis=1)
y = df_no_vs_early['has_cancer']
```

```python
# Train test split

x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.33,  
                                                    stratify=y)
```

```python
# Tune hyperparameters

rf_tuned = tune_hyperparameters(x_train, y_train, RandomForestClassifier(), param_space_rf)
```

```python
show_conf_mat(rf_tuned, ['Cancer','No cancer'],'Confusion Matrix (Stage I/II Cancer vs. No Cancer):', fig_no=2)
```

The accuracy score here is very misleading because the dataset is extremely unbalanced, and the score is strongly influenced by the correct classifications of no cancer. 

However, we notice that the recall score is quite bad. On the one hand, this may suggest that urine biomarkers cannot successfuly identify early stage pancreatic cancer, and this method of diagnosis is ultimately useless. 

On the other hand, we must realize that this dataset is extremely unbalanced with very few stage I and II cancer samples. One of the issues is that PDAC is a rare cancer, and being caught at an early stage is extremely unlikely. This makes collection of early stage cancer samples quite difficult. In the event that a significant number of urine samples from patients with early stage cancer are obtained, this method merits a revisit. 

## 5.2 Compare 'Control' vs. 'Benign'

Another possibility is that benign pancreatic issues show similarities with early stage cancer, and may be confounding the results, since they are grouped with the control group. This can be tested in two ways. First attempt to classify control vs. benign samples. If this is reasonably successful, attempt to classify benign vs. early stage cancer samples.

Quite a good recall score. Clearly, it is possible to identify control samples vs. benign samples. Next we see if we can distinguish between benign and early stage cancer samples.
